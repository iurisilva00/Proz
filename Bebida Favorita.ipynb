##Almoço preferido

```python
prato_preferido = input("Escreva seu Prato Favorito:")
preco = float(input("Escreva o valor do seu prato:"))
orcamento =float(input("Escreva seu orcamento:"))
bebida = float(input("Escreva o valor da bebibda:"))
percentual = 0.1
amigo =input("Tenho um bom amigo {Sim ou Não}:")



valorTotal_prato = preco+bebida
valor_garcom=valorTotal_prato*percentual
```

```python
valor_total = valor_garcom+valorTotal_prato
print(f"Meu prato preferido é {prato_preferido}, o preço do prato é R${preco:.2f}, tomei uma bebida do valor de R${bebida:.2f}  o valor do garçom foi de R${valor_garcom:.2f}\nMeu valor total a pagar foi R${valor_total:.2f}\nO meu orçamento foi de R${orcamento:.2f}")
```

```python

if (orcamento>=valor_total):
  print("Paguei a conta")
elif amigo == True:
  print("Amigo Pagou a conta")

else:
    print("Lavar os pratos")
```
