```python
#Aqui definiremos a função
def calculadora (num1, num2, operador=None):
  if operador =="+":
    return num1+num2
  elif operador =="-":
    return num1-num2
  elif operador =="*":
    return num1*num2
  elif operador =="/":
    return num1/num2
  else:
    return 0

#Inserção dos dados
#Definindo se o usuário deseja continuar ou sair


sair = ""
while sair.lower() != "sair":
  numero_1 = float(input("Digite seu primeiro número:" ))
  numero_2 = float(input("Digite seu segundo número:" ))
  operador = (input("Digite seu operador:" ))


  resposta = calculadora(numero_1, numero_2, operador)

  print("Sua Resposta é:",resposta)
  sair = input("Digite sair para sair do fluxo: ")
```
