```python
# Precisamos imprimir um número para cada andar de um hotel de 20 andares. Porém, o dono do hotel é supersticioso e optou por não ter um 13ro andar.

# Escreva um código que imprima todos os números exceto o número 13.
# Escreva mais um código que resolva o mesmo problema, mas dessa vez usando o laço de repetição 'while'.

# Como desafio, imprima eles em ordem decrescente (20, 19, 18...)
```

```python
#Hotel tem 20 andares, vamos imprir cada um dos número, pulando o décimo terceiro andar, usando laço for
hotel = [] # Nesse caso crio uma lista chamada hotel
for andar in range(1,21):
  if andar == 13:
    continue
  else:
    hotel.append(andar) # Aqui adiciono os andares na minha lista hotel

hotel.sort(reverse=True) # Reverto a ordem da lista

for andar in hotel: # Esse for percorrerá a lista e imprimirá cada valor
  print(f"Esse é o Andar:{andar}")
```

```python
# Usando laço while
hotel = [] # Crio a lista hotel
andar = 1 # Esse contador irá inserir os andares no hotel
while andar <= 20:
  if andar == 13:
      andar += 1
      continue
  hotel.append(andar)
  andar += 1
  hotel.sort(reverse=True) #Reverte a ordem da lista

andar = 0 # Nessse caso usei o contador iniciando em zero, pois o contador de lista inicia em zero
while andar < len(hotel):
    print("Esse é o andar", hotel[andar])
    andar += 1
```
